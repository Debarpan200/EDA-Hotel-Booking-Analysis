# **Project Summary -**

This analysis delved into a comprehensive dataset of 119,390 hotel bookings, aiming to uncover insights into customer behavior and booking patterns. The dataset encompassed a wide range of variables, including booking details, customer demographics, room preferences, and cancellation information.

Before diving into the analysis, the dataset underwent a rigorous cleaning process. Duplicate records were eliminated, and missing values were addressed. The 'country', 'children', and 'agent' variables were imputed with their respective modes, while the 'company' variable, due to its high percentage of missing values, was dropped. 

The EDA phase involved a combination of statistical analysis and data visualization techniques:

* Descriptive Statistics: 
Key metrics such as mean, median, mode, standard deviation, and quartiles were calculated for numerical variables.

* Data Visualization:
1. Bar Plots: Compared categorical variables against numerical variables to identify trends.
2. Heatmaps: Visualized the correlation matrix to understand the relationships between numerical variables.
3. Pair Plots: Explored pairwise relationships between some numerical variables.
4. Line Plots + Scatter Plot: line plots are used in showing the trands/count over time period and Scatterplot is used in specifing the data points.

Main Libraries to be used:

1. Pandas for data manipulation, aggregation
2. Matplotlib and Seaborn for visualisation and behaviour with respect to the target variable.
3. NumPy for computationally efficient operations

## **Let's Start Visualizing**

### **Question 1:** What is the total number of hotel booked Yearly?

```python
data = read_hotel_booking['arrival_date_year'].value_counts(sort=True)
```

```python
sns.barplot(x=data.index, y=data.values)
plt.title('Total Hotel Booking Yearly')
plt.xlabel('Years')
plt.ylabel('Booking')
plt.show()
```

The maximum booking Yearly is in 2016 with approx 40000 bookings as per the DataSet

### **Question 2:** Total Monthly Booking in 2015, 2016 & 2017?

```python
desired_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

#### Monthly Booking in 2015

```python
df_2015 = read_hotel_booking[read_hotel_booking['arrival_date_year'] == 2015]

# Count the number of bookings for each month in 2015
month_counts = df_2015['arrival_date_month'].value_counts().reindex(desired_order).dropna()

print(month_counts)
```

```python
sns.lineplot(x=month_counts.index, y=month_counts.values)
sns.scatterplot(x=month_counts.index, y=month_counts.values)
plt.title('Total Hotel Booking Monthly in 2015')
plt.xlabel('Month')
plt.ylabel('Booking')
plt.ylim(1600)
plt.show()
```

In 2015 from July till December - 
* Maximum number of booking is in September with approx 2800 bookings 
* Minimum number of booking in July & Novemver with Approx 1600 bookings. 

#### Monthly Booking in 2016

```python
df_2016 = read_hotel_booking[read_hotel_booking['arrival_date_year'] == 2016]


month_counts_1 = df_2016['arrival_date_month'].value_counts().reindex(desired_order)
```

```python
sns.lineplot(x=month_counts_1.index, y=month_counts_1.values)
sns.scatterplot(x=month_counts_1.index, y=month_counts_1.values)
plt.title('Total Hotel Booking Monthly in 2016')
plt.xlabel('Month')
plt.ylabel('Booking')
plt.xticks(rotation=90)
plt.ylim(1500)
plt.show()
```

In 2016 - 
* Maximum number of booking is in August with approx 4400 bookings 
* Minimum number of booking in January with Approx 1800 bookings. 

#### Monthly Booking in 2017

```python
df_2017 = read_hotel_booking[read_hotel_booking['arrival_date_year'] == 2017]

# Count the number of bookings for each month in 2015
month_counts_2 = df_2017['arrival_date_month'].value_counts().reindex(desired_order).dropna()
```

```python
sns.lineplot(x=month_counts_2.index, y=month_counts_2.values)
sns.scatterplot(x=month_counts_2.index, y=month_counts_2.values)
plt.title('Total Hotel Booking Monthly in 2017')
plt.xlabel('Month')
plt.ylabel('Booking')
plt.ylim(2750)
plt.show()
```

In 2017 from January till August - 
* Maximum number of booking is in May with approx 4500 bookings

### **Question 5:** Which Hotel type is the Busiest?

### **Question 6:** Count the total reservation canceled in each type of Hotel?


```python
df_ResortHotel = read_hotel_booking[read_hotel_booking['hotel'] == 'Resort Hotel']

Resort_cancel_counts = (df_ResortHotel['is_canceled'] == 1).sum()

print('Total number of Resort Hotel Booking Canceled:',Resort_cancel_counts)
```

```python
df_CityHotel = read_hotel_booking[read_hotel_booking['hotel'] == 'City Hotel']

City_cancel_counts = (df_CityHotel['is_canceled'] == 1).sum()

print('Total number of City Hotel Booking Canceled:',City_cancel_counts)
```

```python
hotel_types = ['Resort Hotel', 'City Hotel']
cancel_counts = [Resort_cancel_counts, City_cancel_counts]
```

```python
sns.barplot(x = hotel_types, y = cancel_counts)
plt.xlabel('Hotel Type')
plt.ylabel('Number of Cancellations')
plt.title('Hotel Booking Cancellations by Type of Hotel')
plt.show()
```

City Hotel have maximum number of booking cancelled in the DataSet
